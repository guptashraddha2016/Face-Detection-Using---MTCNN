```python
import mtcnn
```

```python
detector = mtcnn.MTCNN()
```

```python
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread("/content/Jiya.jpg")
```

```python
face = detector.detect_faces(image)
for box in face:
  box = box['box']
  cv2.rectangle(image,(box[0],box[1]),(box[0]+box[2],box[1]+box[3]),(0,0,0),1)
cv2.imwrite('Detected.jpg',image) 
cv2_imshow(image)
```
